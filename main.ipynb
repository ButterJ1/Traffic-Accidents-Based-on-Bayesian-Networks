```python
import pandas as pd
import csv

def calculate_percentages(series):
    frequency = series.value_counts().to_dict()
    total = series.shape[0]
    percentages = {num: (count / total) * 100 for num, count in frequency.items()}
    
    adjusted_percentages = {num: round(pct, 1) for num, pct in percentages.items()}
    
    total_adjusted = sum(adjusted_percentages.values())
    
    if total_adjusted != 100:
        diff = 100 - total_adjusted
        max_key = max(adjusted_percentages, key=adjusted_percentages.get)
        adjusted_percentages[max_key] += diff
    
    return frequency, adjusted_percentages

csv_data = []

for column in data.columns:
    frequency, adjusted_percentages = calculate_percentages(data[column])
    header = [column] + [f'{val}_frequency' for val in sorted(frequency.keys())] + [f'{val}_percentage' for val in sorted(adjusted_percentages.keys())]
    rows = ["頻率"] + [frequency[val] for val in sorted(frequency.keys())] + [adjusted_percentages[val] for val in sorted(adjusted_percentages.keys())]
    csv_data.append((header, rows))

with open('priors_table.csv', 'w', newline='', encoding='utf-8-sig') as file:
    writer = csv.writer(file)
    for header, rows in csv_data:
        writer.writerow(header)
        writer.writerow(rows)

print("CSV文件已生成：priors_table.csv")
```

def calculate_conditional_probabilities(data, relationships):
    conditional_probabilities = {}
    for parents, child in relationships:
        if not isinstance(parents, tuple):
            parents = (parents,)
        
        # 計算 P(A, B, C, ...)
        parent_counts = data.groupby(list(parents)).size().div(len(data))
        
        # 計算 P(A, B, C, ... ∩ 子)
        joint_counts = data.groupby(list(parents) + [child]).size().unstack().fillna(0)
        joint_probabilities = joint_counts.div(len(data))
        
        # 計算 P(子 | A, B, C, ...)
        conditional_prob = joint_probabilities.div(parent_counts, axis=0)
        
        # 儲存條件機率
        conditional_probabilities[(parents, child)] = conditional_prob
    return conditional_probabilities

def calculate_probabilities(data, columns):
    probabilities = {}
    for column in columns:
        value_counts = data[column].value_counts(normalize=True)
        probabilities[column] = value_counts
    return probabilities

```python
import itertools
import pandas as pd

relationships = [
    (('天候',), '路面狀態'),
    (('車輛撞擊部位1', '當事者行動狀態', '路面狀態'), '主要傷處'),
    (('主要傷處',), '事故位置'),
    (('主要傷處',), '速限-速度限制'),
    (('主要傷處',), '事故類型及型態'),
    (('主要傷處',), '受傷程度'),
    (('受傷程度',), '行動電話'),
    (('受傷程度',), '保護裝備'),
    (('受傷程度',), '道路照明設備'),
    (('受傷程度',), '肇事逃逸'),
    (('肇事逃逸',), '飲酒情形'),
    (('肇事逃逸',), '駕照狀態'),
]

def calculate_conditional_probabilities(data, relationships):
    result_dfs = []
    
    all_columns = set(sum([list(features) for features, target in relationships], []) + 
                      [target for _, target in relationships])
    missing_columns = all_columns - set(data.columns)
    if missing_columns:
        raise KeyError(f"Missing columns in data: {missing_columns}")
    
    for features, target_feature in relationships:
        feature_names = list(features)
        feature_combinations = list(itertools.product(*[sorted(data[feature].unique()) for feature in features]))
        
        # total_samples = len(data) 
        joint_probabilities = data.groupby(list(features)).size().to_dict()
        
        conditional_probabilities = {}
        target_values = sorted(data[target_feature].unique())
        for combination in feature_combinations:
            key = combination if len(features) > 1 else combination[0]
            conditional_probabilities[key] = {}
            subset = data
            for feature, value in zip(features, combination):
                subset = subset[subset[feature] == value]
            for value in target_values:
                count_target = len(subset[subset[target_feature] == value])
                count_condition = joint_probabilities.get(key, 0)
                conditional_probabilities[key][value] = count_target / count_condition if count_condition > 0 else 0
        
        index_names = feature_names
        columns = [f"{target_feature}={value}" for value in target_values]
        result_data = []
        for combination in feature_combinations:
            row = list(combination)
            result_row = []
            key = combination if len(features) > 1 else combination[0]
            for value in target_values:
                result_row.append(conditional_probabilities[key].get(value, 0))
            
            total = sum(result_row)
            if total > 0:
                result_row = [round(x / total * 100, 2) for x in result_row]
            else:
                result_row = [0] * len(result_row)
            
            result_row = [round(x, 1) for x in result_row]
            total = sum(result_row)
            if total != 100:
                max_index = result_row.index(max(result_row))
                result_row[max_index] += 100 - total
                result_row[max_index] = round(result_row[max_index], 1)
            
            result_data.append(row + result_row)
        
        result_df = pd.DataFrame(result_data, columns=index_names + columns)
        result_dfs.append(result_df)
    
    return result_dfs

try:
    conditional_probabilities_dfs = calculate_conditional_probabilities(data, relationships)

    for i, df in enumerate(conditional_probabilities_dfs):
        print(f"\nConditional probabilities for relationship {i + 1}:")
        print(df)
        # Save to CSV file
        csv_filename = f'conditional_probabilities_{i + 1}.csv'
        df.to_csv(csv_filename, index=False)
        print(f"Conditional probabilities saved to {csv_filename}")

    print("All calculations and saves completed.")
except KeyError as e:
    print(f"Error: {e}")
```
